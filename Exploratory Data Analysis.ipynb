```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
%matplotlib inline
```

```python
df = pd.read_csv("E:\DA project\Software professionals survey\m2_survey_data.csv")
```

# Distribution

```python
plt.figure(figsize = (10,5))
sns.distplot(a=df["ConvertedComp"], bins = 20, hist = False)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df["Age"], kde = True);
```

```python
sns.kdeplot(df["ConvertedComp"]);
```

```python
# Find out the median ConvertedComp of responders identified themselves only as a Woman?
woman = df[df['Gender'] == 'Woman']
woman['ConvertedComp'].median()
```

```python
# calculate a 5-number summary
data=np.array(df['Age'])
# calculate quartiles
quartiles = np.nanpercentile(data, [25, 50, 75])
# calculate min/max
data_min, data_max = df['Age'].min(), df['Age'].max()
# print 5-number summary
print('Min: %.3f' % data_min)
print('Q1: %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % data_max)
```

```python
df.hist(column = 'Age', bins=range(0,80,8), figsize=(10,5), color='#86bf91', zorder=2, rwidth=0.9);
```

```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1
# lấy kết quả làm tròn đến chữ số thứ 3
print('Inter-Quartile Range: %.3f' % IQR)
```

```python
LBound = Q1 - 1.5*IQR
UBound = Q1 + 1.5*IQR
print('Upper bound: %.3f' % UBound)
print('Lower bound: %.3f' % LBound)
```
